```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
cols=['temp','Dry_Cough','Fatigue','Loss_Of_Appetite','Sore_Throat','Headache','Loss_Of_Smell/Taste','Vomiting']
for i in cols:
    df[i] = label_encoder.fit_transform(df[i])
df.head()
```

```python
df['Body_Aches']=df['Body_Aches'].replace(['no','normal','high','extreme','extreme '],[0,1,2,3,3])
df['Body_Aches'].unique()
```

```python
df['Shortness_Of_Breath']=df['Shortness_Of_Breath'].replace(['no','normal','high','extreme','extreme '],[0,1,2,3,3])
df['Shortness_Of_Breath'].unique()
```

```python
df
```

```python
features=df[['temp','Dry_Cough','Fatigue','Loss_Of_Appetite','Body_Aches','Shortness_Of_Breath','Sore_Throat','Headache','Loss_Of_Smell/Taste','Vomiting']]
label=df['src_node']
```

```python
from collections import Counter


def euclidean_distance(x1, x2):
        return np.sqrt(np.sum((x1 - x2)**2))
    
def manhattan_distance(x1, x2):
        return np.sum(np.abs(x1 - x2))

# cosine similarity between A and B
def cosine_distance(x1,x2):
    return np.dot(x1,x2)/(np.linalg.norm(x1)*np.linalg.norm(x2))

def hamming_distance(x1,x2):
    i = 0
    count = 0
 
    while(i < len(x1)):
        if(x1[i] != x2[i]):
            count += 1
        i += 1
    return count


class KNN:

    def __init__(self, k=1):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        y_pred = [self._predict(x) for x in X]
        return np.array(y_pred)

    def _predict(self, x):
        # Compute distances between x and all examples in the training set
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
        # Sort by distance and return indices of the first k neighbors
        k_idx = np.argsort(distances)[:self.k]
        # Extract the labels of the k nearest neighbor training samples
        k_neighbor_labels = [self.y_train[i] for i in k_idx]  
        # return the most common class label
        most_common = Counter(k_neighbor_labels).most_common(1)
        return most_common[0][0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
distance_metrics=['manhattan','euclidean','hamming','cosine']

for i in distance_metrics:
    model = KNeighborsClassifier(metric=i,n_neighbors=1)

    # Train the model using the dataset
    model.fit(features,label)

    #Predict Output
    predicted= model.predict([[0,1,1,1,3,2,1,1,1,1]])
    print(predicted)
```
